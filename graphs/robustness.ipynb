```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Visualization setup
sns.set_theme(style="white", context="talk")
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

def load_robustness_results(condition_name, path_pattern, levels, results_dir='../results/robustness/'):
    all_data = []
    
    for level in levels:
        filepath = f'{results_dir}{path_pattern}{level}'
        print(filepath)
        try:
            df = pd.read_csv(filepath)
            df['condition_level'] = float(level)
            all_data.append(df)
        except FileNotFoundError:
            print("NOT FOUND")
            continue
    
    if not all_data:
        raise ValueError(f"No data files found for {condition_name} in {results_dir}")
    
    combined_df = pd.concat(all_data, ignore_index=True)
    return combined_df

def plot_robustness_metrics_for_condition(condition_name, df):
    metrics_config = [
        ('overall_robustness', 'Overall Robustness', 1.0, '#d62728'),
        ('stability', 'Stability (WHY)\nPrediction Consistency', 1.0, '#1f77b4'),
        ('resilience', 'Resilience (HOW)\nAccuracy Retention', 1.0, '#ff7f0e'),
        ('reliability', 'Reliability (WHAT)\nCalibration (1-ECE)', 1.0, '#2ca02c')
    ]
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    axes = axes.flatten()
    
    fig.suptitle(f'Robustness Analysis: {condition_name}', 
                 fontsize=24, fontweight='bold', y=0.98)
    
    df_sorted = df.sort_values('condition_level')
    
    for idx, (col, label, ideal_value, color) in enumerate(metrics_config):
        ax = axes[idx]
        x = df_sorted['condition_level']
        y = df_sorted[col]
        
        ax.axhline(ideal_value, color='gray', linestyle='--', linewidth=1.5, alpha=0.5, label='Ideal')
        ax.plot(x, y, marker='o', markersize=10, linewidth=3, color=color, label=label)
        ax.fill_between(x, y, ideal_value, color=color, alpha=0.1)
        
        for i in [y.index[0], y.index[-1]]: 
            val = df_sorted.loc[i, col]
            lvl = df_sorted.loc[i, 'condition_level']
            ax.annotate(f'{val:.3f}', xy=(lvl, val), xytext=(0, 10), 
                        textcoords='offset points', ha='center', fontweight='bold', color=color)
        
        ax.set_title(label, fontsize=14, fontweight='bold')
        ax.set_xlabel('Bias/Noise Level', fontsize=12)
        ax.set_ylabel('Score', fontsize=12)
        
        y_min = max(0, y.min() - 0.05)
        y_max = 1.02
        ax.set_ylim(y_min, y_max)
        ax.grid(axis='y', linestyle=':', alpha=0.4)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    return fig
def plot_comparative_robustness(conditions, levels, results_dir='../results/robustness/'):
    all_conditions_data = []

    for cond in conditions:
        try:
            df = load_robustness_results(cond['name'], cond['pattern'], levels, results_dir)
            df['Condition_Type'] = cond['name']
            all_conditions_data.append(df)
        except Exception as e:
            print(f"Skipping {cond['name']} for comparison: {e}")

    if not all_conditions_data:
        print("No robustness data found to compare.")
        return

    full_df = pd.concat(all_conditions_data, ignore_index=True)
    full_df['condition_level'] = full_df['condition_level'].astype(float)
    full_df = full_df.sort_values(['Condition_Type', 'condition_level'])

    metrics_to_plot = [
        ('overall_robustness', 'Overall Robustness'),
        ('stability', 'Stability'),
        ('resilience', 'Resilience'),
        ('reliability', 'Reliability')
    ]

    fig, axes = plt.subplots(2, 2, figsize=(20, 15))
    axes = axes.flatten()
    palette = sns.color_palette("tab10", len(conditions))
    line_styles = ['-', '--', '-.', ':', '-', '--', '-.', ':', '-', '--']
    markers = ['o', 's', '^', 'D', 'v', 'p', '*', 'X', 'P', 'h']

    for idx, (col, title) in enumerate(metrics_to_plot):
        ax = axes[idx]
        
        ax.axhline(1.0, color='darkgreen', linestyle='--', linewidth=2.5, alpha=0.8, label='Ideal', zorder=1)
        condition_final_values = []
        for i, cond in enumerate(conditions):
            cond_data = full_df[full_df['Condition_Type'] == cond['name']]
            if len(cond_data) > 0 and col in cond_data.columns:
                final_value = cond_data[col].iloc[-1]
                condition_final_values.append((i, cond, abs(final_value - 1.0)))

        condition_final_values.sort(key=lambda x: x[2])
        for plot_idx, (i, cond, _) in enumerate(condition_final_values):
            cond_data = full_df[full_df['Condition_Type'] == cond['name']]
            if len(cond_data) > 0:
                ax.plot(cond_data['condition_level'], cond_data[col],
                       marker=markers[i % len(markers)],
                       markersize=7,
                       linewidth=2.5,
                       linestyle=line_styles[i % len(line_styles)],
                       label=cond['name'],
                       color=palette[i],
                       alpha=0.75,
                       zorder=2 + plot_idx)
        
        ax.set_title(f'Comparison: {title}', fontsize=16, fontweight='bold', pad=15)
        data_min = full_df[col].min()
        data_max = full_df[col].max()
        data_range = data_max - data_min

        margin = max(data_range * 0.1, 0.02)
        y_min = max(0, data_min - margin)
        y_max = min(1.05, data_max + margin)

        if 1.0 > y_max:
            y_max = min(1.05, 1.0 + 0.05)
        
        ax.set_ylim(y_min, y_max)
        
        ax.set_xlabel('Condition Level', fontsize=13, labelpad=10)
        ax.set_ylabel('Score', fontsize=13, labelpad=10)
        ax.grid(True, linestyle=':', alpha=0.3, zorder=0)
        ax.tick_params(labelsize=11)

        if idx == 0:
            handles, labels = ax.get_legend_handles_labels()
            if 'Ideal' in labels:
                ideal_idx = labels.index('Ideal')
                ideal_handle = handles.pop(ideal_idx)
                ideal_label = labels.pop(ideal_idx)
                
                sorted_pairs = sorted(zip(labels, handles))
                labels, handles = zip(*sorted_pairs)
                labels = list(labels)
                handles = list(handles)
                
                labels.insert(0, ideal_label)
                handles.insert(0, ideal_handle)
            
            ax.legend(handles, labels,
                     bbox_to_anchor=(1.05, 1), loc='upper left',
                     fontsize=9, title='Bias Condition',
                     title_fontsize=10, framealpha=0.95)
        else:
            if ax.get_legend():
                ax.get_legend().remove()

    plt.suptitle('Global Robustness Comparison Across All Bias Conditions', 
                 fontsize=24, fontweight='bold', y=0.98)
    
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    print("="*80)
    print("ROBUSTNESS VISUALIZATION - ALL CONDITIONS")
    print("="*80)
    
    levels = ['0.00', '0.10', '0.20', '0.30', '0.40', '0.50', '0.60', '0.70', '0.80', '0.90', '1.00']
    conditions = [
        {'name': 'Hist. Bias on Y','pattern': 'hist_bias_Y_ly_'},
        {'name': 'Hist. Bias on Q','pattern': 'hist_bias_Q_lhq_'},
        {'name': 'Hist. Bias on R','pattern': 'hist_bias_R_lhr_'},
        {'name': 'Interaction Proxy','pattern': 'interaction_bias_lyb_'},
        {'name': 'Meas. Bias Linear Y','pattern': 'meas_bias_Y_linear_lmy_'},
        {'name': 'Meas. Bias Non Linear Y','pattern': 'meas_bias_Y_nonlinear_lmy_'},
        {'name': 'Meas. Bias on R','pattern': 'meas_bias_R_lm_'},
        {'name': 'Undersampling','pattern': 'undersample_pu_'},
        {'name': 'Representation Bias','pattern': 'representation_bias_lr_true'},
        {'name': 'Label Noise','pattern': 'label_noise_sy_'},
    ]
    
    for condition in conditions:
        try:
            print(f"Processing: {condition['name']}")
            df = load_robustness_results(condition['name'], condition['pattern'], levels)
            
            fig = plot_robustness_metrics_for_condition(condition['name'], df)
            
            plt.show()
        except Exception as e:
            print(f"   Skipping {condition['name']}: {e}")

    plot_comparative_robustness(conditions, levels)
```
