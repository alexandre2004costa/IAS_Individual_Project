```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from pathlib import Path

# Visualization setup
sns.set_theme(style="white", context="talk")
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False

def load_robustness_results(condition_name, path_pattern, levels, results_dir='../results/robustness/'):
    """Load robustness results for a specific condition."""
    all_data = []
    
    for level in levels:
        # Check for common naming variations (e.g. 0.1 vs 0.10)
        filepath = f'{results_dir}{path_pattern}{level}'
        print(filepath)
        try:
            df = pd.read_csv(filepath)
            df['condition_level'] = float(level)
            all_data.append(df)
        except FileNotFoundError:
            print("NOT FOUND")
            continue
    
    if not all_data:
        raise ValueError(f"No data files found for {condition_name} in {results_dir}")
    
    combined_df = pd.concat(all_data, ignore_index=True)
    return combined_df

def plot_robustness_metrics_for_condition(condition_name, df):
    """Plot all 4 Robustness metrics for a single condition with ZOOMED view."""
    
    # Define metrics based on your Robustness Pipeline output
    metrics_config = [
        ('overall_robustness', 'Overall Robustness', 1.0, '#d62728'),
        ('stability', 'Stability (WHY)\nPrediction Consistency', 1.0, '#1f77b4'),
        ('resilience', 'Resilience (HOW)\nAccuracy Retention', 1.0, '#ff7f0e'),
        ('reliability', 'Reliability (WHAT)\nCalibration (1-ECE)', 1.0, '#2ca02c')
    ]
    
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    axes = axes.flatten()
    
    fig.suptitle(f'Robustness Analysis: {condition_name}', 
                 fontsize=24, fontweight='bold', y=0.98)
    
    df_sorted = df.sort_values('condition_level')
    
    for idx, (col, label, ideal_value, color) in enumerate(metrics_config):
        ax = axes[idx]
        x = df_sorted['condition_level']
        y = df_sorted[col]
        
        # 1. Ideal reference line
        ax.axhline(ideal_value, color='gray', linestyle='--', linewidth=1.5, alpha=0.5, label='Ideal')
        
        # 2. Main metric line
        ax.plot(x, y, marker='o', markersize=10, linewidth=3, color=color, label=label)
        
        # 3. Deviation fill
        ax.fill_between(x, y, ideal_value, color=color, alpha=0.1)
        
        # 4. Annotations
        for i in [y.index[0], y.index[-1]]: # Start and End points
            val = df_sorted.loc[i, col]
            lvl = df_sorted.loc[i, 'condition_level']
            ax.annotate(f'{val:.3f}', xy=(lvl, val), xytext=(0, 10), 
                        textcoords='offset points', ha='center', fontweight='bold', color=color)
        
        # 5. Styling
        ax.set_title(label, fontsize=14, fontweight='bold')
        ax.set_xlabel('Bias/Noise Level', fontsize=12)
        ax.set_ylabel('Score', fontsize=12)
        
        # 6. ZOOM Logic
        y_min = max(0, y.min() - 0.05)
        y_max = 1.02
        ax.set_ylim(y_min, y_max)
        ax.grid(axis='y', linestyle=':', alpha=0.4)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    return fig
def plot_comparative_robustness(conditions, levels, results_dir='../results/robustness/'):
    """
    Plots a 2x2 comparison where each subplot contains lines for ALL conditions.
    This shows which bias types impact robustness the most.
    """
    all_conditions_data = []

    # 1. Aggregate data from all conditions into one DataFrame
    for cond in conditions:
        try:
            df = load_robustness_results(cond['name'], cond['pattern'], levels, results_dir)
            df['Condition_Type'] = cond['name']
            all_conditions_data.append(df)
        except Exception as e:
            print(f"Skipping {cond['name']} for comparison: {e}")

    if not all_conditions_data:
        return

    full_df = pd.concat(all_conditions_data, ignore_index=True)
    full_df = full_df.sort_values(['Condition_Type', 'condition_level'])

    # 2. Setup Plotting
    metrics_to_plot = [
        ('overall_robustness', 'Overall Robustness'),
        ('stability', 'Stability'),
        ('resilience', 'Resilience'),
        ('reliability', 'Reliability')
    ]

    fig, axes = plt.subplots(2, 2, figsize=(20, 15))
    axes = axes.flatten()
    
    # Use a distinct color palette for the different bias types
    palette = sns.color_palette("husl", len(conditions))

    for idx, (col, title) in enumerate(metrics_to_plot):
        ax = axes[idx]
        
        sns.lineplot(
            data=full_df, 
            x='condition_level', 
            y=col, 
            hue='Condition_Type', 
            marker='o', 
            ax=ax, 
            palette=palette,
            linewidth=2.5,
            markersize=8
        )
        
        ax.set_title(f'Comparison: {title}', fontsize=16, fontweight='bold')
        ax.set_ylim(max(0, full_df[col].min() - 0.05), 1.02)
        ax.set_xlabel('Bias/Noise Level')
        ax.set_ylabel('Score')
        ax.grid(True, linestyle=':', alpha=0.6)
        
        # Only show legend on the first plot to avoid clutter, 
        # or move it to a dedicated space
        if idx != 0:
            ax.get_legend().remove()
        else:
            ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

    plt.suptitle('Global Robustness Comparison Across All Bias Conditions', fontsize=24, fontweight='bold', y=0.98)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    # Save comparison plot
    output_path = Path("../results/plots/robustness/global_comparison.png")
    output_path.parent.mkdir(parents=True, exist_ok=True)
    plt.savefig(output_path, bbox_inches='tight')
    print(f" Comparative visualization saved to {output_path}")
    plt.show()


if __name__ == "__main__":
    print("="*80)
    print("ROBUSTNESS VISUALIZATION - ALL CONDITIONS")
    print("="*80)
    
    levels = ['0.00', '0.10', '0.20', '0.30', '0.40', '0.50', '0.60', '0.70', '0.80', '0.90', '1.00']
    # Define all conditions to visualize
    conditions = [
        {'name': 'Historical Bias on Y (Label)','pattern': 'hist_bias_Y_ly_'},
        {'name': 'Historical Bias on Q (Feature)','pattern': 'hist_bias_Q_lhq_'},
        {'name': 'Historical Bias on R (Feature)','pattern': 'hist_bias_R_lhr_'},
        {'name': 'Interaction Bias','pattern': 'interaction_bias_lyb_'},
        {'name': 'Meas Linear LMY','pattern': 'meas_bias_Y_linear_lmy_'},
        {'name': 'Meas Non Linear LMY','pattern': 'meas_bias_Y_nonlinear_lmy_'},
        {'name': 'Measurement Bias on R','pattern': 'meas_bias_R_lm_'},
        {'name': 'Undersampling (Imbalance)','pattern': 'undersample_pu_'},
        {'name': 'representation_bias_lr_true','pattern': 'representation_bias_lr_true'},
        {'name': 'Label Noise','pattern': 'label_noise_sy_'},
    ]
    
    for condition in conditions:
        try:
            print(f"Processing: {condition['name']}")
            df = load_robustness_results(condition['name'], condition['pattern'], levels)
            
            fig = plot_robustness_metrics_for_condition(condition['name'], df)
            
            # Save the plot
            output_dir = Path("../results/plots/robustness/")
            output_dir.mkdir(parents=True, exist_ok=True)
            fig.savefig(output_dir / f"{condition['pattern']}plot.png")
            
            plt.show()
        except Exception as e:
            print(f"   Skipping {condition['name']}: {e}")

    plot_comparative_robustness(conditions, levels)
```
