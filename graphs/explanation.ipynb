```python
"""
Detailed DoX Visualization - Individual Metrics per Condition
==============================================================

Creates separate visualizations for each bias type (like fairness pipeline):
- Historical Bias on Y
- Historical Bias on Q  
- Historical Bias on R
- Undersampling
- Label Noise
- etc.

Shows all 4 DoX metrics for each condition.
"""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pathlib import Path

# Visualization setup
sns.set_theme(style="white", context="talk")
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False


def load_dox_results(condition_name, path_pattern, levels, results_dir='../results/explanation/'):
    """
    Load DoX results for a specific condition.
    
    Parameters:
    -----------
    condition_name : str
        Name of condition (e.g., "Historical Bias Y")
    path_pattern : str
        Pattern for file paths (e.g., "hist_bias_Y_ly_{}")
    levels : list
        List of levels to load
    results_dir : str
        Directory containing results
        
    Returns:
    --------
    pd.DataFrame : Combined results
    """
    all_data = []
    
    for level in levels:

        filepath = f'{results_dir}{path_pattern}{level}.csv'
        print(filepath)
        try:
            df = pd.read_csv(filepath)
            df['condition_level'] = level
            all_data.append(df)
        except FileNotFoundError:
            print(f"✗ Not found: {filepath}")
            continue
    
    if not all_data:
        raise ValueError(f"No data files found for {condition_name}")
    
    combined_df = pd.concat(all_data, ignore_index=True)
    
    # Filter to non-reference group (the group being compared)
    if 'group' in combined_df.columns:
        # Assuming Group_0 is reference, Group_1 is comparison
        combined_df = combined_df[combined_df['group'] == 'Group_1']
    
    print(f"✓ Loaded {len(all_data)} levels for {condition_name}")
    print(f"✓ Total rows: {len(combined_df)}")
    
    return combined_df


def plot_dox_metrics_for_condition(condition_name, df):
    """
    Plot all 4 DoX metrics for a single condition with ZOOMED view.
    
    Parameters:
    -----------
    condition_name : str
        Name to display in title
    df : pd.DataFrame
        Results dataframe with condition_level column
    """
    # Define metrics with ideal values and colors
    metrics_config = [
        ('mean_dox', 'Overall DoX Score', 1.0, '#2c7bb6'),
        ('mean_clarity', 'Clarity (WHY)\nTop Features Identified', 1.0, '#fdae61'),
        ('mean_distinctiveness', 'Distinctiveness (HOW)\nRelative Importance', 1.0, '#d7191c'),
        ('mean_coverage', 'Coverage (WHAT)\nTop Features Explain', 1.0, '#91bfdb')
    ]
    
    # Create figure with 2x2 subplots
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    axes = axes.flatten()
    
    # Main title
    fig.suptitle(f'DoX Metrics: {condition_name}', 
                 fontsize=24, fontweight='bold', y=0.98)
    
    # Sort by level
    df_sorted = df.sort_values('condition_level')
    
    # Plot each metric
    for idx, (col, label, ideal_value, color) in enumerate(metrics_config):
        ax = axes[idx]
        
        # Extract data
        x = df_sorted['condition_level']
        y = df_sorted[col]
        
        # 1. Draw ideal reference line
        ax.axhline(ideal_value, color='green', linestyle='--', linewidth=2.5, 
                   alpha=0.7, zorder=1, label='Ideal Value')
        
        # 2. Draw the main metric line with markers
        ax.plot(x, y, marker='o', markersize=12, linewidth=3.5, 
                color=color, label=label, zorder=2)
        
        # 3. Fill area to show deviation from ideal
        ax.fill_between(x, y, ideal_value, color=color, alpha=0.2)
        
        # 4. Annotate key points (first, last, worst)
        idx_min = y.idxmin()
        points_to_annotate = [y.index[0], y.index[-1]]
        if idx_min not in points_to_annotate:
            points_to_annotate.append(idx_min)
        
        for i in points_to_annotate:
            val = df_sorted.loc[i, col]
            level_val = df_sorted.loc[i, 'condition_level']
            ax.annotate(f'{val:.3f}', xy=(level_val, val), 
                       xytext=(0, 10), textcoords='offset points',
                       ha='center', fontsize=11, fontweight='bold', 
                       color=color, bbox=dict(boxstyle='round,pad=0.4', 
                       facecolor='white', edgecolor=color, alpha=0.8))
        
        # 5. Styling
        ax.set_title(label, fontsize=14, pad=12, fontweight='bold')
        ax.set_xlabel('Condition Level', fontsize=12, labelpad=10)
        ax.set_ylabel('Score', fontsize=12, labelpad=10)
        ax.grid(axis='y', linestyle=':', alpha=0.4)
        
        # 6. ZOOM: Dynamic y-limits based on data range
        data_range = y.max() - y.min()
        data_margin = max(data_range * 0.15, 0.02)  # At least 15% margin or 0.02
        
        # Set limits but ensure we show the ideal value if it's close
        y_min = max(0, min(y.min() - data_margin, ideal_value - 0.05))
        y_max = min(1.05, max(y.max() + data_margin, ideal_value + 0.05))
        
        ax.set_ylim(y_min, y_max)
        
        # Add legend for first plot
        if idx == 0:
            ax.legend(fontsize=11, loc='best')
    
    plt.tight_layout()
    return fig

def plot_comparative_dox(conditions, levels, results_dir='../results/explanation/'):
    """
    Plots a 2x2 comparison for Explanation (DoX) metrics across all conditions.
    """
    all_conditions_data = []

    # 1. Aggregate DoX data
    for cond in conditions:
        try:
            # Reusing your loading logic: filters for Group_1 comparison
            df = load_dox_results(cond['name'], cond['pattern'], levels, results_dir)
            df['Condition_Type'] = cond['name']
            all_conditions_data.append(df)
        except Exception as e:
            print(f"Skipping {cond['name']} for DoX comparison: {e}")

    if not all_conditions_data:
        print("No DoX data found to compare.")
        return

    full_df = pd.concat(all_conditions_data, ignore_index=True)
    full_df['condition_level'] = full_df['condition_level'].astype(float)
    full_df = full_df.sort_values(['Condition_Type', 'condition_level'])

    # 2. Setup Plotting for the 4 DoX metrics
    metrics_to_plot = [
        ('mean_dox', 'Overall DoX Score'),
        ('mean_clarity', 'Clarity (WHY)'),
        ('mean_distinctiveness', 'Distinctiveness (HOW)'),
        ('mean_coverage', 'Coverage (WHAT)')
    ]

    fig, axes = plt.subplots(2, 2, figsize=(20, 15))
    axes = axes.flatten()
    
    palette = sns.color_palette("bright", len(conditions))

    for idx, (col, title) in enumerate(metrics_to_plot):
        ax = axes[idx]
        
        sns.lineplot(
            data=full_df, 
            x='condition_level', 
            y=col, 
            hue='Condition_Type', 
            marker='o', 
            ax=ax, 
            palette=palette,
            linewidth=2.5,
            markersize=8
        )
        
        ax.set_title(f'Comparison: {title}', fontsize=16, fontweight='bold')
        
        # Adjust Y-axis to show decay clearly
        y_min = max(0, full_df[col].min() - 0.05)
        ax.set_ylim(y_min, 1.05)
        
        ax.set_xlabel('Bias Level')
        ax.set_ylabel('Score')
        ax.grid(True, linestyle=':', alpha=0.6)
        
        # Legend management
        if idx != 0:
            ax.get_legend().remove()
        else:
            ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

    plt.suptitle('Global Explanation (DoX) Comparison Across All Bias Conditions', 
                 fontsize=24, fontweight='bold', y=0.98)
    
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    # Save to the explanation plots folder
    output_path = Path("../results/plots/explanation/global_dox_comparison.png")
    output_path.parent.mkdir(parents=True, exist_ok=True)
    plt.savefig(output_path, bbox_inches='tight')
    
    print(f"✓ Comparative DoX visualization saved to {output_path}")
    plt.show()
# =============================================================================
# MAIN EXECUTION
# =============================================================================

if __name__ == "__main__":
    """Generate all DoX visualizations."""
    print("="*80)
    print("DETAILED DOX VISUALIZATION - ALL CONDITIONS")
    print("="*80)
    levels = ['0.00', '0.10', '0.20', '0.30', '0.40', '0.50', '0.60', '0.70', '0.80', '0.90', '1.00']
    # Define all conditions to visualize
    conditions = [
        {'name': 'Historical Bias on Y (Label)','pattern': 'hist_bias_Y_ly_'},
        {'name': 'Historical Bias on Q (Feature)','pattern': 'hist_bias_Q_lhq_'},
        {'name': 'Historical Bias on R (Feature)','pattern': 'hist_bias_R_lhr_'},
        {'name': 'Interaction Bias','pattern': 'interaction_bias_lyb_'},
        {'name': 'Meas Linear LMY','pattern': 'meas_bias_Y_linear_lmy_'},
        {'name': 'Meas Non Linear LMY','pattern': 'meas_bias_Y_nonlinear_lmy_'},
        {'name': 'Measurement Bias on R','pattern': 'meas_bias_R_lm_'},
        {'name': 'Undersampling (Imbalance)','pattern': 'undersample_pu_'},
        {'name': 'representation_bias_lr_true','pattern': 'representation_bias_lr_true'},
        {'name': 'Label Noise','pattern': 'label_noise_sy_'},
    ]
    
    # Process each condition
    for i, condition in enumerate(conditions, 1):
        print(f"\n{'='*80}")
        print(f"[{i}/{len(conditions)}] Processing: {condition['name']}")
        print('='*80)
        
        try:
            # Load data
            print(f"\nLoading data for {condition['name']}...")
            df = load_dox_results(
                condition['name'],
                condition['pattern'],
                levels
            )
            
            # Create metrics plot
            print(f"\nCreating DoX metrics plot...")
            fig1 = plot_dox_metrics_for_condition(condition['name'], df)
            
            # Show plots
            plt.show()
            
            print(f"✓ Completed: {condition['name']}")
            
        except Exception as e:
            print(f"✗ Error processing {condition['name']}: {e}")
            import traceback
            traceback.print_exc()
            continue
    
    print("Comparison plot:")
    plot_comparative_dox(conditions, levels)
```
